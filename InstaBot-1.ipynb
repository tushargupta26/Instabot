## 5 Like/Unlike Posts

###  5.1 Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
driver.get('https://www.instagaram.com')
driver.implicitly_wait(10)
search = driver.find_element_by_class_name('XTCLo')
search.clear()
search.send_keys('dilsefoodie')
link = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]//a')))
link.click()
time.sleep(3)
driver.find_element_by_xpath('//div[contains(@class, "Nnq7C")]/div/a/div[1]').click()
j=0
while True:
    time.sleep(2)
    like_button = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//span[@class = 'fr66n']/button")))
    type_button = BeautifulSoup(like_button.get_attribute('innerHTML'),'html.parser').find('svg').get('aria-label')
    if type_button == 'Like':
        like_button.click()
        print(j+1,' post liked')
    else:
        print(j+1,' post was already liked')
    j+=1
#     break
    if j==30:
        break
    try:
        next1 = driver.find_element_by_class_name('coreSpriteRightPaginationArrow')
        next1.click()
    except:
        print('No More posts by user')
```

```python
# Function to like any no. of posts of any user
def like(user_name,total_posts_liked):
    driver.get('https://www.instagaram.com')
    driver.implicitly_wait(10)
    search = driver.find_element_by_class_name('XTCLo')
    search.clear()
    search.send_keys(user_name)
    # driver.implicitly_wait(10)
    link = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]//a')))
    link.click()
    time.sleep(3)
    driver.find_element_by_xpath('//div[contains(@class, "Nnq7C")]/div/a/div[1]').click()
    j=0
    while True:
        time.sleep(2)
        like_button = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//span[@class = 'fr66n']/button")))
        type_button = BeautifulSoup(like_button.get_attribute('innerHTML'),'html.parser').find('svg').get('aria-label')
        if type_button == 'Like':
            like_button.click()
            print(j+1,' post liked')
        else:
            print(j+1,' post was already liked')
        j+=1
        if j==total_posts_liked:
            break
        try:
            next1 = driver.find_element_by_class_name('coreSpriteRightPaginationArrow')
            next1.click()
        except:
            print('No More posts by user')
            break
            
like('dilsefoodie',30)
```

   #### Or The user has no story.
   #### Or View the story if not yet seen.


```python
driver.get('https://www.instagaram.com')
search = driver.find_element_by_class_name('XTCLo')
search.clear()
search.send_keys('coding.ninjas')
# driver.implicitly_wait(10)
link = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//div[@class="fuqBx"]//a')))
link.click()
time.sleep(3)
try:
    if int(driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]/canvas').get_attribute('height'))==210:
        print('Story not seen Showing it now')
        driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]').click()
    elif int(driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]/canvas').get_attribute('height'))==208:
        print('Story Already Seen!')
except:
    print('The user has no story!')
```
